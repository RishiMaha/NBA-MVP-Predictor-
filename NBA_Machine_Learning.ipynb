### Error Metrics

```python
# Identifying error metric
def find_ap(combined):
    actual = combined.sort_values("Share", ascending=False).head(5)
    predicted = combined.sort_values("Predictions", ascending=False)
    pts = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            pts.append(found/seen)
        seen += 1
    return sum(pts) / len(pts)
```

```python
find_ap(combined)
```

```python
years = list(range(2000, 2025))
```

```python
# Calculating top 5 MVPs for every year
aps = []
all_predictions = []
for year in years[5:]:
    train = stats[stats["Year"] <year]
    test = stats[stats["Year"] == year]
    regression.fit(train[predictors], train["Share"])
    predictions = regression.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["Predictions"], index=test.index)
    combined = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combined)
    aps.append(find_ap(combined))
    
```

```python
# Mean average precision
sum(aps) / len(aps)
```

### Rank & Differential

```python
# Add ranks for MVP voting 
def add_ranks(combined):
    combined = combined.sort_values("Share", ascending=False)
    combined["Rk"] = list(range(1, combined.shape[0] + 1))
    combined = combined.sort_values("Predictions", ascending=False)
    combined["Predicted_Rk"] = list(range(1, combined.shape[0] + 1))
    combined["Diff"] = combined["Rk"] - combined["Predicted_Rk"]
    return combined 
```

```python
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"] < 6].sort_values("Diff", ascending=False)
```

```python
# Backtesting function to get predictions
def backtest(stats, model, year, predictors):
    aps = []
    all_predictions = []
    for year in years[5:]:
        train = stats[stats["Year"] <year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors], train["Share"])
        predictions = regression.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["Predictions"], index=test.index)
        combined = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combined = add_ranks(combined)
        all_predictions.append(combined)
        aps.append(find_ap(combined))
    return sum(aps)/len(aps), aps, pd.concat(all_predictions) 
```

```python
mean_ap, aps, all_predictions = backtest(stats, regression, years[5:], predictors)
```
