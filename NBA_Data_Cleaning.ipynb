### Cleaning the MVP Dataset

### Cleaning the Players Dataset

```python
# Loading in the players dataset
players = pd.read_csv("Player_Stat.csv")
players
```

```python
# Deleting unwanted columns
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head(50)
```

```python
# Removing "*" next to player names

players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
# Player with multiple rows
players.groupby(["Player", "Year"])
```

```python
# Function to change certain players who have multiple rows to a single row 
def one_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players = players.groupby(["Player", "Year"]).apply(one_row)
```

```python
players.head(20)
```

```python
# Drop index levels
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Etdrick Bohannon"]
```

### Merging Player and MVP Datasets

```python
# Merging both datasets
combination = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combination.head(7)
```

```python
# Selecting only players whose Pts won are greater than 0
combination[combination["Pts Won"] > 0]
```

```python
# Replacing any "N/A" values with 0 
combination[["Pts Won", "Pts Max", "Share"]] = combination[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combination
```

### Cleaning Team Data

```python
# Read in teams dataset
teams = pd.read_csv("Teams.csv")
teams
```

```python
teams.head(30)
```

```python
# Remove Rows with "Division" in it & teams with "*" next to it
teams = teams[~teams["W"].str.contains("Division")]
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
import re 



# Define a function to clean the team names
def clean_team_name(teams):
    # Use regex to remove the `\xa0(number)` part
    clean_name = re.sub(r'\xa0\(\d+\)', '', teams)
    return clean_name.strip()

# Apply the cleaning function to the 'Team' column
teams['Team'] = teams['Team'].apply(clean_team_name)
teams
```

```python
combination["Tm"].unique()
```

```python
# Add a column to help map betwee team abbreviation and full names using a dictionary
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combination["Tm"].map(nicknames)
```

```python
nicknames
```

```python
combination["Team"] = combination["Tm"].map(nicknames)
```

```python
combination.head(5)
```

```python
# Adding in the extra columns from the team dataset
stats = combination.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
combination
```

```python
# Remove "Unnamed: 0"
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

```python
# Change data types to numeric
stats = stats.apply(pd.to_numeric, errors="ignore")
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
# Change Games Back column to replace "-" with 0
stats["GB"] = stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
